```python
from treeswift import read_tree_newick
```

```python
trees = read_tree_newick('edit-all-genes_21_200.phylip_phyml_tree.txt')
```

```python
print(type(trees))
print(len(trees))
print(trees[0])
```

```python
f = open("pruned-genes_21_200.phylip.txt", 'w')
for tree in trees:
    tree.contract_low_support(0.05)
    tree.suppress_unifurcations()
    f.write(tree.newick() + '\n')
f.close()
```
