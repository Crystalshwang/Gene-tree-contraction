```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
rf200 = pd.read_excel("rf-200.xlsx")
rf400 = pd.read_excel("rf-400.xlsx")
```

```python
rf200.head()
```

```python
xlabel1 = [0, 5, 10, 20]
ylabel1 = [sum(rf200[0])/len(rf200),sum(rf200[5])/len(rf200),sum(rf200[10])/len(rf200),sum(rf200[20])/len(rf200)]

xlabel2 = [0, 5, 10, 20]
ylabel2 = [sum(rf400[0])/len(rf400),sum(rf400[5])/len(rf400),sum(rf400[10])/len(rf400),sum(rf400[20])/len(rf400)]
```

```python
fig = plt.figure()
ax= plt.subplot(111)
ax.plot(xlabel1, ylabel1, label='200bp')
ax.plot(xlabel2, ylabel2, label='400bp')
plt.title('Average RF distance for different alignments')
plt.xlabel('Contract Threshold')
plt.ylabel('Average RF distance')
ax.legend()
```

```python
fig = plt.figure()
ax= plt.subplot(111)
ax.scatter(np.zeros(len(rf200)), rf200[0],c='r',label='200bp contract = 0')
ax.scatter(np.ones(len(rf200))*5, rf200[5],c='b',label='200bp contract = 5')
ax.scatter(np.ones(len(rf200))*10, rf200[10],c='g',label='200bp contract = 10')
ax.scatter(np.ones(len(rf200))*20, rf200[20],c='c',label='200bp contract = 20')
ax.scatter(np.zeros(len(rf400)), rf400[0],c='m',label='400bp contract = 0')
ax.scatter(np.ones(len(rf400))*5, rf400[5],c='y',label='400bp contract = 5')
ax.scatter(np.ones(len(rf400))*10, rf400[10],c='k',label='400bp contract = 10')
ax.scatter(np.ones(len(rf400))*20, rf400[20],c='#005511',label='400bp contract = 20')

plt.title('RF distance of all replicates with different bp')
plt.xlabel('Contract Threshold')
plt.ylabel('RF distance')
ax.legend()
```
